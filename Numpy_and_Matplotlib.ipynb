**MATPLOTLIB**

```python
x1 = np.arange(0,10) # take 1-10
x2 = np.linspace(0,math.pi/2,101) # create 100 points between 0 and 1 for the graph to make it smoother and more accurate
x3 = np.linspace(0,math.pi/2,101)
y1 = np.arange(10) # take 1-10
y2 = np.sin(x2) # plot a sin curve for the x values
y3 = np.cos(x3)
plt.plot(x2,y2,label="sin(µ)") # plots two arrays
plt.plot(x3,y3,label="cos(µ)") # plot another curve
plt.xlabel("µ") # label x axis
plt.ylabel("sin(µ)") # label y axis
plt.title("µ vs sin(µ) graph") # name the graph
plt.legend() # create a legend for the plot
plt.show() # shows the plot
```

```python
# fig and ax
fig,ax = plt.subplots(1,2)
fig.set_figwidth(20) # adjust the width of the individual plots
fig.set_figheight(20) # adjust the height of the individual plots
# ax is an array of the axes of the different plots
ax
ax[0].scatter(x2,y2)
ax[1].plot(x1,y1)
```
